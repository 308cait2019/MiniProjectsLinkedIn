```python
# Artemis I (Nov 16, 2022) Data
# Using data for the corresponding launch date, time for 2m air temperature
H_artemis1 = Herbie("2022-11-16 01:47", model="hrrr", product="sfc", fxx=0)
ds1 = H_artemis1.xarray("TMP:2 m")
# FLorida Box [W,E,S,N]
fl_extent = [-88, -79, 24, 32]

# Artemis II (Apr 1, 2026) Data
# Using data for the corresponding launch date, time for 2m air temperature
H_artemis2 = Herbie("2026-04-01 22:35", model="hrrr", product="sfc", fxx=0)
ds2 = H_artemis2.xarray("TMP:2 m")
temp2_f = (ds2.t2m - 273.15) * 9/5 + 32
f2_extent = [-88, -79, 24, 32]

# Major Cities and KSC for context
cities = { "Miami": [25.7617, -80.1918],
    "Tampa": [27.9506, -82.4572],
    "Jacksonville": [30.3322, -81.6557],
    "Tallahassee": [30.4383, -84.2807],
    "Kennedy Space Center": [28.5729, -80.6490]}

#print(ds1)
```

```python
# Put cities into pandas format for Herbie to pick points
points_df = pd.DataFrame([
    {"name": city, "latitude": coords[0], "longitude": coords[1]}
    for city, coords in cities.items()
])


# See: https://herbie.readthedocs.io/en/stable/user_guide/tutorial/accessor_notebooks/pick_points.html
# Returns a new, small xarray with points at dictionary listed cities instead of massive weather map
extracted1 = ds1.herbie.pick_points(points_df, method="nearest")
extracted2 = ds2.herbie.pick_points(points_df, method="nearest")

print(f"{'City':<25} | {'Artemis I (°F)':<15} | {'Artemis II (°F)':<15}")
print("-" * 60)

# Create dictionary for values
city_temps = {}

# Loop through the cities to get ast at each point
for i in range(len(points_df)):
    # For city at each row number selected with iloc
    city_name = points_df.iloc[i]["name"]
    
    # Selects positional temperature at each point in points_df
    val1 = extracted1.t2m.isel(point=i).values
    val2 = extracted2.t2m.isel(point=i).values
    
    f1 = (val1 - 273.15) * 9/5 + 32
    f2 = (val2 - 273.15) * 9/5 + 32

    city_temps[city_name] = {
        "Artemis I": round(f1, 2),
        "Artemis II": round(f2, 2) }
    print(f"{city_name:<25} | {f1:>15f} | {f2:>15f}")

print(city_temps)

# Create the grid variable before the plot
temp1_f = (ds1["t2m"] - 273.15) * 9/5 + 32
temp2_f = (ds2["t2m"] - 273.15) * 9/5 + 32

# Verify the valid time (should match the splashdown date)
print(f"Model Run: {dt}")
print(f"Valid Forecast Time: {ds1.valid_time.values}")
print(f"Highest Temp in Area (F): {temp1_f.max().item():.2f}")
print(f"Valid Forecast Time: {ds2.valid_time.values}")
print(f"Highest Temp in Area (F): {temp2_f.max().item():.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
from herbie import Herbie

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
# Artemis I (Nov 16, 2022) Data
# Using data for the corresponding launch date, time for pressure level based data
H_artemis1 = Herbie("2022-11-16 01:47", model="hrrr", product="prs", fxx=0)

# Artemis II (Apr 1, 2026) Data
# Using data for the corresponding launch date, time for pressure level based data
H_artemis2 = Herbie("2026-04-01 22:35", model="hrrr", product="prs", fxx=0)

# Artemis II (Apr 10, 2026) Splashdown Data
# Run from 10:00 UTC, looking 2 days ahead
dt3 = pd.Timestamp("2026-04-08 00:00")
fxx = 48
# Artemis II Splashdown (Apr 10, 2026) Data
H = Herbie(dt3, model="hrrr", product="prs", fxx=fxx)

# Kennedy Space Center Launch location (subtract abs val of lon from 360 to match longitudinal scale 0-360)
# Use pick points method
location = pd.DataFrame({
    "latitude": [28.608],
    "longitude": [279.396]  # or -80.604, Herbie usually handles both here
})

# Extract directional wind components (U and V grid components)
# and geopotential height for pressure levels (HGT)
# for up to ~40,000 ft (200mb) (0-9 is used to ensure all pressure levels of x digits in a row - wildcard)
search_pattern = ":(UGRD|VGRD|HGT):((250|300|500|700|850|925|1000))"

# Extract variables
ds = H_artemis1.xarray(search_pattern)
ds2 = H_artemis2.xarray(search_pattern)
ds3 = H.xarray(search_pattern)

# Merge the list into a single Dataset
if isinstance(ds, list):
    ds_list = xr.merge(ds)
else:
    ds_list = ds

if isinstance(ds2, list):
    ds2_list = xr.merge(ds2)
else:
    ds2_list = ds2

if isinstance(ds3, list):
    ds3_list = xr.merge(ds3)
else:
    ds3_list = ds3
    
# Extract this data for a single point location
pt_data = ds_list.herbie.pick_points(location, method="nearest")
pt_data2 = ds2_list.herbie.pick_points(location, method="nearest")
pt_data3 = ds3_list.herbie.pick_points(location, method="nearest")

# View results
print(pt_data.to_dataframe()[['u', 'v', 'gh']])
print(pt_data2.to_dataframe()[['u', 'v', 'gh']])
print(pt_data3.to_dataframe()[['u', 'v', 'gh']])

# Create u,v winds and geopotential height (pressure above mean sea level)
u_winds = pt_data.u
v_winds = pt_data.v
geopotential_h = pt_data
#print(u_winds)

u_winds2 = pt_data2.u
v_winds2 = pt_data2.v
geopotential_h2 = pt_data2

u_winds3 = pt_data3.u
v_winds3 = pt_data3.v
geopotential_h3 = pt_data3

# Convert to df for calculations, sort by altitude
# remove 0 value point column with .reset_index(level='point', drop=True)
df = pt_data.to_dataframe()[['u', 'v', 'gh']].reset_index(level='point', drop=True)
df = df.sort_values('gh')

df2 = pt_data2.to_dataframe()[['u', 'v', 'gh']].reset_index(level='point', drop=True)
df2 = df2.sort_values('gh')
df3 = pt_data3.to_dataframe()[['u', 'v', 'gh']].reset_index(level='point', drop=True)
df3 = df3.sort_values('gh')

print(df)
print(df2)
print(df3)
```

```python
# WIND SHEAR CALCULATIONS

# Calculate vector wind shear 
# after finding difference between u and v wind and geopotential height between each pressure level
df = df.sort_values('gh')
df['du'] = df['u'].diff()
df['dv'] = df['v'].diff()
df['dz'] = df['gh'].diff()
df['vector_shear_mag'] = np.sqrt(df['du']**2 + df['dv']**2)
df['shear_intensity'] = df['vector_shear_mag'] / df['dz']

df2 = df2.sort_values('gh')
df2['du'] = df2['u'].diff()
df2['dv'] = df2['v'].diff()
df2['dz'] = df2['gh'].diff()
df2['vector_shear_mag'] = np.sqrt(df2['du']**2 + df2['dv']**2)
df2['shear_intensity'] = df2['vector_shear_mag'] / df2['dz']

df3 = df3.sort_values('gh')
df3['du'] = df3['u'].diff()
df3['dv'] = df3['v'].diff()
df3['dz'] = df3['gh'].diff()
df3['vector_shear_mag'] = np.sqrt(df3['du']**2 + df3['dv']**2)
df3['shear_intensity'] = df3['vector_shear_mag'] / df3['dz']

# Order calulated and extracted variables into df, remove first row NaN values as there's no difference
analysis = df[['gh', 'u', 'v', 'vector_shear_mag', 'shear_intensity']].dropna()
analysis2 = df2[['gh', 'u', 'v', 'vector_shear_mag', 'shear_intensity']].dropna()
analysis3 = df3[['gh', 'u', 'v', 'vector_shear_mag', 'shear_intensity']].dropna()

print(analysis3)
```

```python
# CAPE

# Define the launch windows as starting Midnight UTC and ending 24 hours later on launch dates
run_date = "2022-11-16 00:00" 
run_date2 = "2026-04-01 00:00" 
forecast_range = range(0, 24)  # Forecast hours 0 through 24

# Kennedy Space Center Launch location
# Use pick points method
location = pd.DataFrame({
    "latitude": [28.608],
    "longitude": [-80.604]
})

data_list = []
data_list2 = []

for f in forecast_range:
    H = Herbie(run_date, model="hrrr", product="sfc", fxx=f)
    ds = H.xarray(":CAPE:surface")
    H2 = Herbie(run_date2, model="hrrr", product="sfc", fxx=f)
    ds2 = H2.xarray(":CAPE:surface")
    
    # Extract the point for KSC
    pt = ds.herbie.pick_points(location, method="nearest")
    data_list.append(pt.to_dataframe())
    pt2 = ds2.herbie.pick_points(location, method="nearest")
    data_list2.append(pt2.to_dataframe())

# Combine into one DataFrame for plotting
df_plot = pd.concat(data_list)
df_plot2 = pd.concat(data_list2)
df_plot2
```

```python
# SURFACE LIFTING INDEX

data_list3 = []
data_list4 = []

for f in forecast_range:
    # --- Process 2022 Date ---
    H3 = Herbie(run_date, model="hrrr", product="prs", fxx=f)
    ds_grid3 = H3.xarray(":(?:4?LFTX):")
    
    # Merge if multiple GRIB messages (LFTX and 4LFTX) are found
    if isinstance(ds_grid3, list):
        ds_grid3 = xr.merge(ds_grid3)
    
    pt3 = ds_grid3.herbie.pick_points(location, method="nearest")
    data_list3.append(pt3.to_dataframe())

    # --- Process 2026 Date ---
    H4 = Herbie(run_date2, model="hrrr", product="prs", fxx=f)
    ds_grid4 = H4.xarray(":(?:4?LFTX):")
    
    if isinstance(ds_grid4, list):
        ds_grid4 = xr.merge(ds_grid4)
        
    pt4 = ds_grid4.herbie.pick_points(location, method="nearest")
    data_list4.append(pt4.to_dataframe())

# Combine into one DataFrame for plotting
df_lftx_2022 = pd.concat(data_list3)
df_lftx_2026 = pd.concat(data_list4)

# Conversion function (Celsius to Fahrenheit)
def c_to_f(celsius):
    return (celsius * 9/5) + 32

# Apply to the columns in both dataframes
# Note: Using '4lftx' and 'lftx' as these are the standard Herbie names
for df in [df_lftx_2022, df_lftx_2026]:
    if '4lftx' in df.columns:
        df['4lftx'] = df['4lftx'].apply(c_to_f)
    if 'lftx' in df.columns:
        df['lftx'] = df['lftx'].apply(c_to_f)

# Display result for 2026
df_lftx_2026
```
